```python
import pandas as pd
import os
from openpyxl import load_workbook

def extract_tables_from_workbooks(input_dir, output_dir):
    # Create the output directory if it does not exist
    if not os.path.exists(output_dir):
        os.makedirs(output_dir)

    # Define the strings with leading and trailing spaces
    start_string = 'UNITED HEALTHCARE UNDERWRITING FULLY INSURED APPROVAL FORM - For Internal Use Only'
    end_string = 'GRAND TOTAL'

    # Iterate through each file in the directory
    for filename in os.listdir(input_dir):
        if not filename.startswith("~$") and filename.endswith(".xlsx"): # Ensure that only Excel files are considered and exclude temporary files
            file_path = os.path.join(input_dir, filename)
            xls = pd.ExcelFile(file_path)
            
            # Iterate through each sheet in the Excel file
            for sheet_name in xls.sheet_names:
                print(f"Processing sheet: {sheet_name} in file: {filename}")
                
                # Read the sheet into a DataFrame skipping header and footer rows
                df = pd.read_excel(file_path, sheet_name=sheet_name, header=None, skipfooter=1, dtype=str)
                
                # Initialize flags to identify the start and end of the table
                table_started = False
                
                # Iterate through rows in the DataFrame
                for index, row in df.iterrows():
                    # Check if the start string is found in the row
                    if start_string in ' '.join(map(str, row)):
                        start_row_index = index
                        table_started = True
                    # Check if the end string is found in the row and the table has started
                    elif table_started and end_string in ' '.join(map(str, row)):
                        end_row_index = index
                        # Select the table rows between start and end row indexes
                        selected_table = df.iloc[start_row_index:end_row_index + 1]
                    
                        # Save the selected table into a new Excel file with the workbook name
                        output_file_path = os.path.join(output_dir, f"{filename}")
                        selected_table.to_excel(output_file_path, index=False, header=False)
                        print(f"Extracted table saved to: {output_file_path}")
                        
                        # Reset the flag
                        table_started = False
                        # Break the loop as we found the end of the table
                        break

# Example usage:
input_dir = r"C:\Users\shres\Downloads\Start_End_Worksheetsststd"
output_dir = r"C:\Users\shres\Downloads\Extracted_Tables_Tuftsdddddddfffhhhkkkff"
extract_tables_from_workbooks(input_dir, output_dir)
```
